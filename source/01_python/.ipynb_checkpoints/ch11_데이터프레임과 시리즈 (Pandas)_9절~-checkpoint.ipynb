```python
# ★ 2차원 데이터 프레임의 apply 이용하는 방법
```

```python
def year_month_append(df):
    'df를 받아 년도와 월 추가후 return'
    df = df.copy() # 깊은 복사
    df['년'] = int(df['년도월'].split('년')[0])
    df['월'] = int(df['년도월'].split('년')[1].split('월')[0])
    return df
```

```python
new_df = df_mltd.apply(year_month_append, axis=1)
new_df.head()
```

# 12절. 일괄변경하기(결측치, 특정값)
## 12.1 fillna
- 대부분 결측치는 drop하거나, 평균값대체, 중위수 대체, 그룹별 평균값 혹은 중위수, 인접(이전, 이후) 값 등으로 대체


```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                  [3, 4, np.nan, 1],
                  [np.nan, 3, np.nan, 5],
                  [np.nan, 3, np.nan, 4]],
                columns=[list('ABCD')])
df
```

```python
# 모든 결측치를 0 대체
df.fillna(0)
```

```python
# 결측치를 이전행의 값으로 대체
df.fillna(method='ffill')
```

```python
# 결측치를 다음행의 값으로 대체
df.fillna(method='bfill')
```

```python
# 결측치를 열마다 특정값으로 대체
values = {'A':85, 'B':80, 'C':75, 'D':90}
df.fillna(value=values)
```

```python
# 결측치를 열별 평균값으로 대체
df.fillna(value=df.mean(axis=0))
```

```python
# 결측치 대체를 2번째 결측치까지만 대체
df.fillna(value=values, limit=2)
```

### ★ ★ ★예제 (80page) ★ ★ ★

```python
import random
import numpy as np
import pandas as pd
import seaborn as sns
iris = sns.load_dataset('iris')
rowlen, collen = iris.shape
rowlen, collen
```

```python
# (1단계) iris 각 열마다 20개씩 인위적으로 결측치를 할당한다.
print('임의의 20개 index 추출 :',random.sample(range(rowlen), 20))
```

```python
# 결측치 할당
for col in range(collen-1):
    iris.iloc[random.sample(range(rowlen), 20), col] = np.nan
```

```python
# 결측치 확인 - iris.isna(), iris.isnull():결측치인지 여부
# iris.isna().sum()
print(iris.isnull().sum())
iris.tail()
```

```python
# 결측치가 하나라도 있는 행 출력
iris[iris['sepal_length'].isna() | iris['sepal_width'].isna() | 
     iris['petal_length'].isna() | iris['petal_width'].isna()]
```

```python
# (2-1단계) 결측치는 각 열의 평균으로 대체한다 : fillna사용
#iris.mean(numeric_only=True)
#iris.iloc[:, :-1].mean()
iris.fillna(iris.iloc[:, :-1].mean())
```

```python
iris.isna().sum()
```

```python
# (2-2단계) 결측치는 종별 일별 중위수(gropuby, pivot_table)로 대체한다 : fillna 사용x
iris_gr_avg = iris.groupby('species').mean()
iris_gr_avg
```

```python
# fillna사용X 확인
iris.fillna(iris_gr_avg) 
```

```python
iris.iloc[1]
```

```python
iris_gr_avg # iris_gr_avg.loc['setosa','sepal_length']
```

```python
# iris 한행을 매개변수로 받아 결측치가 있는 경우 종별 열별 중위수로 대체하는 함수
def iris_fillna(iris_row):
    iris_row = iris_row.copy()
    for col, value in iris_row.items():
        if isinstance(value, (int, float, np.number)) and np.isnan(value):
            # print(irisrow[col], '이 결측치')
            iris_row[col] = iris_gr_avg.loc[iris_row.species, col]
    return iris_row
```

```python
iris_fillna(iris.iloc[2])
```

```python
# apply 적용
iris = iris.apply(iris_fillna, axis=1)
iris
```
