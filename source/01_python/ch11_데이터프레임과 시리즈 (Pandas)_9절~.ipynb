```python
# 연습하기(apply, map)
# df_mltd의 연도월 데이터를 연도와 월을 int로 분리하여 컬럼을 추가하세요.
df_mltd.head(2)
```

```python
df_mltd['Year'] = df_mltd['년도월'].apply(lambda datestr : int(datestr.split('년')[0]))
#df_mltd['Month'] = df_mltd['년도월'].apply(lambda datestr : datestr.split('년')[1][:-2])
df_mltd['Month'] = df_mltd['년도월'].apply(lambda datestr : int(datestr.split('년')[1].split('월')[0]))
df_mltd.head(2)
```

```python
# ★ 2차원 데이터 프레임의 apply 이용하는 방법
```

```python
def year_month_append(df):
    'df를 받아 년도와 월 추가후 return'
    df = df.copy() # 깊은 복사
    df['년'] = int(df['년도월'].split('년')[0])
    df['월'] = int(df['년도월'].split('년')[1].split('월')[0])
    return df
```

```python
new_df = df_mltd.apply(year_month_append, axis=1)
new_df.head()
```

# 12절. 일괄변경하기(결측치, 특정값)
## 12.1 fillna
- 대부분 결측치는 drop하거나, 평균값대체, 중위수 대체, 그룹별 평균값 혹은 중위수, 인접(이전, 이후) 값 등으로 대체


```python
df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                  [3, 4, np.nan, 1],
                  [np.nan, 3, np.nan, 5],
                  [np.nan, 3, np.nan, 4]],
                columns=[list('ABCD')])
df
```

```python
# 모든 결측치를 0 대체
df.fillna(0)
```

```python
# 결측치를 이전행의 값으로 대체
df.fillna(method='ffill')
```

```python
# 결측치를 다음행의 값으로 대체
df.fillna(method='bfill')
```

```python
# 결측치를 열마다 특정값으로 대체
values = {'A':85, 'B':80, 'C':75, 'D':90}
df.fillna(value=values)
```

```python
# 결측치를 열별 평균값으로 대체
df.fillna(value=df.mean(axis=0))
```

```python
# 결측치 대체를 2번째 결측치까지만 대체
df.fillna(value=values, limit=2)
```
